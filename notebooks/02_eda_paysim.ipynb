6.3M transactions. Focus: fraud type breakdown, imbalance, temporal patterns.

```python
import sys; sys.path.append('..')
import pandas as pd
import matplotlib.pyplot as plt
from src.data.paysim_loader import load_raw, filter_fraud_types, get_label_column

df = load_raw()

print(df.shape)
print(df['type'].value_counts())
fraud_df = filter_fraud_types(df)
print('\nFraud rate in TRANSFER+CASH_OUT:', fraud_df[get_label_column()].mean())
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fraud_c = fraud_df['type'].value_counts()
fraud_c.plot(kind='bar', ax=axes[0])
axes[0].set_title('Fraud Type Counts')
axes[0].set_xlabel('Fraud Type')
axes[0].set_ylabel('Count')
non_fraud_c = df['type'].value_counts()
non_fraud_c.plot(kind='bar', ax=axes[1])
axes[1].set_title('Non-Fraud Type Counts')
axes[1].set_xlabel('Non-Fraud Type')
axes[1].set_ylabel('Count')



```
